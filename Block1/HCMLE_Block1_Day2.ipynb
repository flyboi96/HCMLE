```python
'''
1. FizzBuzz Agent Filter

Write a function agent_filter(n) that prints:
	•	"Fizz" if the number is divisible by 3,
	•	"Buzz" if divisible by 5,
	•	"FizzBuzz" if divisible by both,
	•	The number itself otherwise.

Then:
Apply it to all integers from 1 to 100 using a loop. Print the results with line numbers like a classified report.
'''

def agent_filter(n):
    div_3 = n%3 == 0
    div_5 = n%5 == 0
    if div_3 and div_5:
        print('FizzBuzz')
    elif div_3:
        print('Fizz')
    elif div_5:
        print('Buzz')
    else:
        print(f'{n:d}')

for i in range(1,101):
    agent_filter(i)
```

```python
'''
2. Word Frequency Analyzer

Create a .txt file named intercepted_message.txt (or I can generate one for you).

Write functions to:
	•	Count total words
	•	Count unique words
	•	Identify the 5 most common words
	•	Display words sorted by frequency

Use string methods and collections.Counter.
'''

import re

def remove_non_letters(text):
  return re.sub(r'[^a-zA-Z]', '', text)

def iterate_words_in_file(filepath):
    try:
        with open(filepath, 'r') as file:
            word_dict = {}
            for line in file:
                for word in line.split():
                    word_lower = remove_non_letters(word.lower())
                    if len(word_lower) == 0:
                        continue
                    elif word_lower in word_dict:
                        word_dict[word_lower] = word_dict[word_lower] + 1
                    else:
                        word_dict[word_lower] = 1
        return word_dict
    except FileNotFoundError:
        print(f'Error: File not ofund at {filepath}')
    except Exception as e:
        print(f'An error occurred: {e}')

filename = 'intercepted_message.txt'
word_dict = iterate_words_in_file(filename)

total_words = 0
for key in word_dict:
    total_words += word_dict[key]
print(f'Total words: {total_words:d}')

print(f'Unique words: {len(word_dict):d}')

sorted_dict_by_values = sorted(word_dict.items(), key=lambda item: item[1], reverse=True)
first_five_keys = [item[0] for item in sorted_dict_by_values[:5]]
#sorted_dict_by_values
print('Five most common words: ', first_five_keys)

print('Words sorted by frequency:', sorted_dict_by_values)
```

```python
'''
3. Hidden Message Decoder

Write a function that finds words in the text that:
	•	Are exactly 7 characters long
	•	Contain at least one uppercase letter
	•	End in a vowel

Print all that qualify—these may be codenames or message triggers.
'''

import re

def remove_non_letters(text):
  return re.sub(r'[^a-zA-Z]', '', text)

def iterate_words_in_file(filepath):
    try:
        with open(filepath, 'r') as file:
            word_dict = {}
            for line in file:
                for word in line.split():
                    word_only = remove_non_letters(word)
                    if len(word_only) == 0:
                        continue
                    elif word_only in word_dict:
                        word_dict[word_only] = word_dict[word_only] + 1
                    else:
                        word_dict[word_only] = 1
        return word_dict
    except FileNotFoundError:
        print(f'Error: File not ofund at {filepath}')
    except Exception as e:
        print(f'An error occurred: {e}')

filename = 'intercepted_message.txt'
word_dict = iterate_words_in_file(filename)
word_list = list(word_dict.keys())

word_trigger_list = []
the_vowel = ["a","e","i","o","u"]
for word in word_list:
    if len(word) == 7 and any(char.isupper() for char in word) and word[-1].lower() in the_vowel:
        word_trigger_list.append(word)
print('All qualifying words: ', word_trigger_list)
```

```python
'''
4. String Metrics Dashboard

Display:
	•	Average word length
	•	Longest and shortest words
	•	Number of sentences (hint: count . and !)
'''

total_char = sum(len(s) for s in word_list)
total_words = len(word_list)
avg_char = total_char / total_words
print(f'Average word length: {avg_char:.1f}')

shortest_string = min(word_list, key=len)
longest_string = max(word_list, key=len)
print(f'Longest word: {longest_string:s}')
print(f'Shortest word: {shortest_string:s}')

def count_sentences(filepath):
    try:
        with open(filepath, 'r') as file:
            text = file.read()
            sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|!)\s', text)
            return len(sentences)
    except FileNotFoundError:
        return "File not found."

filename = 'intercepted_message.txt'
sentence_count = count_sentences(filename)

if isinstance(sentence_count, int):
    print(f"The number of sentences in '{filename}' is: {sentence_count}")
else:
    print(sentence_count)
```
