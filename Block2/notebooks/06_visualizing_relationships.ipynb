```python
# Session 6: Visualizing Relationships and Comparisons with Seaborn
# Author: Alex Corbin
# Date: 2025-04-23

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

sns.set(style="whitegrid")
os.makedirs("plots", exist_ok=True)
```

```python
titanic = sns.load_dataset("titanic")

plt.figure(figsize=(8, 6))
sns.barplot(data=titanic, x='class', y='survived', estimator=lambda x: sum(x)/len(x))
# sns.barplot(data=titanic, x='class', y='survived', estimator=np.mean)
plt.title('Titanic Survival Rate by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.tight_layout()
plt.savefig('plots/titanic_class_survival_bar.png', dpi=300)
plt.show()
```

# Titanic Survival Rate by Passenger Class – Barplot Analysis

## Purpose
This barplot visualizes the **mean survival rate** for each class of passengers aboard the Titanic. Since the `survived` variable is binary (1 = survived, 0 = did not survive), the average survival value is directly interpretable as a **proportion (rate)**. This allows us to compare outcomes across the ship’s social strata.

---

## Observations

1. **Survival Rate Decreases with Class**

   - **First Class** passengers had the highest survival rate (~63%).
   - **Second Class** survival drops to ~47%.
   - **Third Class** passengers had the lowest chance of survival (~24%).

2. **Strong Stratification by Class**

   - The plot reveals a **clear socioeconomic survival disparity**, supporting historical accounts that upper-deck passengers had more access to lifeboats.

3. **Error Bars Provide Context**

   - The vertical lines represent **confidence intervals** (by default, ~95% CI for the mean).
   - There is **non-overlap between First and Third Class**, reinforcing statistical separation.
   - Slight overlap between First and Second may merit further testing (e.g., t-tests).

---

## Design Notes

- Used a custom `estimator=lambda x: sum(x)/len(x)` to explicitly calculate survival proportion.
- Figure is sized for clarity and saved at 300 dpi for reproducibility.
- Axis labels and title provide immediate interpretability without inspecting the data code.

---

## Learning Objectives Achieved

### **Objective: Visualize group-level comparisons using categorical plots**

- Used `sns.barplot()` with an explicit estimator to compute and compare proportions across categories.
- Interpreted **statistical confidence intervals** and their implications for group differences.
- Practiced encoding meaningful comparisons using visual grammar (length = rate, bars = category, CI = uncertainty).

---

```python
tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 6))
sns.violinplot(data=tips, x='day', y='total_bill', inner='quartile')
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day of Week')
plt.ylabel('Total Bill ($)')
plt.tight_layout()
plt.savefig('plots/tips_day_total_violin.png', dpi=300)
plt.show()
```

# Total Bill Distribution by Day – Violinplot Analysis

## Purpose
This visualization reveals how **total restaurant bills** vary by day of the week. The **violinplot** combines a boxplot's summary statistics with a **kernel density estimate** (KDE), allowing for both visual **distribution shape** and **central tendency** analysis.

---

## Observations

1. **Saturday and Sunday Show Wider Distributions**

   - Both weekend days exhibit **greater spread and higher maximums** in total bill amounts.
   - This suggests higher variability and possibly larger party sizes or longer visits on weekends.

2. **Weekday Distributions Are More Concentrated**

   - Thursday and Friday bills tend to cluster around $12–$20 with a tighter KDE.
   - The quartile lines for weekdays are more compact, indicating **less variance in spending behavior**.

3. **Skewness and Peaks**

   - The KDE tails show moderate **right-skewness** across all days.
   - The **density peaks** suggest common tipping ranges: most meals fall between $10–$20.

---

## Design Notes

- `inner='quartile'` overlays horizontal lines for the **25th, 50th (median), and 75th percentiles**.
- The violin’s width is proportional to **density**, offering an intuitive way to see how concentrated the spending is around certain values.
- Plot was saved at high resolution (300 dpi) for future use in reports or presentations.

---

```python
plt.figure(figsize=(8, 6))
sns.stripplot(data=tips, x='smoker', y='tip', jitter=True, alpha=0.7)
plt.title('Tip Amounts by Smoking Status')
plt.xlabel('Smoker')
plt.ylabel('Tip ($)')
plt.tight_layout()
plt.savefig('plots/tips_smoker_tip_scatter.png', dpi=300)
plt.show()
```

# Tip Amounts by Smoking Status – Stripplot Analysis

## Purpose
This plot compares **individual tip values** between smokers and non-smokers using a **categorical scatterplot**. A `stripplot` is ideal for examining **raw data dispersion** in a single dimension, revealing micro-patterns that summary statistics can obscure.

---

## Observations

1. **High Variability Within Both Groups**

   - Both smokers and non-smokers exhibit **wide vertical spread**, with most tips clustering between $2 and $5.
   - No single group shows consistent dominance in tip amount.

2. **Outliers Are Present in Both Groups**

   - A few non-smokers tipped above $8–$9, with one just under $10.
   - Smokers also had rare high tips (e.g., around $6–$7), though slightly fewer extreme values appear.

3. **Tip Concentration Similar**

   - Both categories have the **highest density of points between $2–$4**, suggesting similar central tendencies.
   - Visual inspection does not suggest a strong relationship between smoking status and tip generosity.

---

## Design Notes

- `jitter=True` prevents overplotting by horizontally offsetting dots within each category.
- `alpha=0.7` improves **dot clarity** in dense areas without overpowering overlapping points.
- Figure size and layout are optimized for visibility and saved to a clean output path.
