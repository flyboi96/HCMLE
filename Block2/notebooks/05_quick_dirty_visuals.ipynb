```python
# Session 5: Quick & Dirty Visuals with Seaborn
# Goal: Learn to visualize trends, patterns, and data signals
# Author: Alex Corbin
# Date: 2025-04-21

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import os

# Create directory if it doesn't exist
os.makedirs('plots', exist_ok=True)

# Set default style
sns.set(style='whitegrid')
```

```python
# Load datasets
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')

# Preview
titanic.head()
iris.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=titanic, x='class', y='fare', hue='sex')
plt.title('Titanic Fare by Class and Gender')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.legend(title='Sex')
plt.tight_layout()
plt.savefig('plots/titanic_fare_boxplot.png', dpi=300)
plt.show()
```

# Titanic Fare by Class and Gender – Boxplot Analysis

## Purpose
This visualization explores how **ticket fares** varied across **passenger class** and **gender** aboard the Titanic. A boxplot is well-suited for this task as it concisely summarizes **distributional characteristics** such as median, interquartile range (IQR), and the presence of outliers.

---

## 📈 Observations

1. **Fare Increases with Class**  

   - Median fares **increase sharply** from Third to First Class.
   - First Class passengers show the **widest spread** in fares, with significant outliers on the high end (above \$500).

2. **Gender Differences Within Classes**  

   - In **First Class**, females tend to have **higher median fares** than males. This may reflect shared cabin costs, group bookings, or gender-based travel patterns.
   - In **Second and Third Class**, fare differences by gender are **minimal**, with males showing slightly lower fare ranges overall.

3. **Outliers and Variability**  

   - All classes display outliers, with **First Class** having the most, indicating **price stratification**.
   - The **narrowest fare distribution** is among **Third Class males**, suggesting constrained pricing in that category.

---

## Design Rationale

- `hue='sex'` provides categorical clarity without overcrowding the plot.
- The **(10, 6)** figure size improves spacing and readability.
- `tight_layout()` and labeled axes ensure the plot is **presentation-ready**.
- The output was saved at **300 dpi** to a structured `plots/` folder for reuse in reports or GitHub.

# Iris Petal Length vs Width by Species – Scatterplot Analysis

## Purpose
This scatterplot explores the relationship between **petal length** and **petal width** across the three species of the Iris dataset: *setosa*, *versicolor*, and *virginica*. The goal is to visualize how these two continuous features interact and whether they can differentiate species through simple visual inspection.

---

## Observations

1. **Distinct Clustering by Species**  

   - *Setosa* (blue) is clearly separated in the bottom-left corner with **short and narrow petals**.
   - *Versicolor* (orange) forms a **moderately linear cluster** in the mid-range of petal dimensions.
   - *Virginica* (green) occupies the upper-right, showing **longer and wider petals**.

2. **Linearity Within Species**  

   - *Versicolor* and *virginica* show a **strong positive correlation** between petal length and width.
   - *Setosa* displays minimal internal variation — a tight, compact cluster.

3. **Potential for Classification**  

   - The plot visually demonstrates why **petal length and width** are highly discriminative features in the classic Iris classification problem.
   - The species are **linearly separable** to a high degree, particularly between *setosa* and the others.

---

## Design Rationale

- Used `sns.scatterplot()` with `hue='species'` to distinguish groups by color.
- Marker size `s=70` balances visibility and density.
- Axis labels include units (cm) for scientific clarity.
- The use of `tight_layout()` ensures clean export for publication or presentation.

```python
plt.figure(figsize=(8,6))
sns.histplot(data=titanic, x='age', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Age on Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('plots/titanic_age_hist.png', dpi=300)
plt.show()
```

# Distribution of Age on Titanic – Histogram with KDE

## Purpose
This histogram visualizes the distribution of passenger ages aboard the Titanic. It provides insight into the demographic makeup of the passengers and reveals patterns that may inform further survival or class-based analyses. A **Kernel Density Estimate (KDE)** is layered on top to provide a smooth approximation of the underlying distribution.

---

## Observations

1. **Age Distribution Skewed Right (Positive Skew)**  

   - The majority of passengers were between **20 and 40 years old**, with a pronounced peak around **25**.
   - The tail extends into older age ranges, but passengers over **60** were relatively few.

2. **Children and Teens** 

   - There is a secondary small bump around **age 0–10**, likely representing children and infants aboard. The drop at age 10 suggests discrete age reporting or cabin assignment patterns.

3. **Continuous Distribution with Outliers**  

   - The KDE curve highlights that the data isn’t uniformly distributed and shows a **long tail** on the right — characteristic of age data in open populations.

---

## Design Rationale

- `bins=30` balances granularity and clarity, avoiding overfitting the noise.
- `kde=True` overlays a smooth distribution to identify shape and modality.
- `color='skyblue'` improves visual softness and accessibility.
- Plot is saved at **300 dpi** for publication-ready quality.
